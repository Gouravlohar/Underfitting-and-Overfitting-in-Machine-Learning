```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

# Generate some synthetic data for the example
np.random.seed(0)
X = np.sort(5 * np.random.rand(80, 1), axis=0)
y = np.sin(X).ravel() + np.random.rand(80) * 0.2

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Linear Regression
linear_regressor = LinearRegression()
linear_regressor.fit(X_train, y_train)
y_linear_pred = linear_regressor.predict(X_test)

# Polynomial Regression with degree 10 (Overfitting)
poly_features = PolynomialFeatures(degree=10)
X_poly = poly_features.fit_transform(X_train)
poly_regressor = LinearRegression()
poly_regressor.fit(X_poly, y_train)
X_test_poly = poly_features.transform(X_test)
y_poly_pred = poly_regressor.predict(X_test_poly)

# Plot the results
plt.scatter(X, y, label='Training Data')
plt.plot(X_test, y_linear_pred, label='Linear Regression', linewidth=2, color='r')
plt.plot(X_test, y_poly_pred, label='Polynomial Regression (Overfitting)', linewidth=2, color='g')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Overfitting and Underfitting Example')
plt.legend()
plt.show()

# Calculate Mean Squared Error for both models
mse_linear = mean_squared_error(y_test, y_linear_pred)
mse_poly = mean_squared_error(y_test, y_poly_pred)
print(f'Mean Squared Error (Linear Regression): {mse_linear:.2f}')
print(f'Mean Squared Error (Polynomial Regression): {mse_poly:.2f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

# Generate some synthetic data for the example
np.random.seed(0)
X = np.sort(5 * np.random.rand(80, 1), axis=0)
y = np.sin(X).ravel() + np.random.rand(80) * 0.2

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Linear Regression
linear_regressor = LinearRegression()
linear_regressor.fit(X_train, y_train)
y_linear_pred = linear_regressor.predict(X_test)

# Polynomial Regression with degree 10 (Overfitting)
poly_features = PolynomialFeatures(degree=10)
X_poly = poly_features.fit_transform(X_train)
poly_regressor = LinearRegression()
poly_regressor.fit(X_poly, y_train)
X_test_poly = poly_features.transform(X_test)
y_poly_pred = poly_regressor.predict(X_test_poly)

# Plot the results side by side
plt.figure(figsize=(12, 5))

# Linear Regression Plot
plt.subplot(1, 2, 1)
plt.scatter(X, y, label='Training Data')
plt.plot(X_test, y_linear_pred, label='Linear Regression', linewidth=2, color='r')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression (Underfitting)')
plt.legend()

# Polynomial Regression Plot
plt.subplot(1, 2, 2)
plt.scatter(X, y, label='Training Data')
plt.plot(X_test, y_poly_pred, label='Polynomial Regression (Overfitting)', linewidth=2, color='g')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression (Overfitting)')
plt.legend()

plt.tight_layout()
plt.show()

# Calculate Mean Squared Error for both models
mse_linear = mean_squared_error(y_test, y_linear_pred)
mse_poly = mean_squared_error(y_test, y_poly_pred)
print(f'Mean Squared Error (Linear Regression): {mse_linear:.2f}')
print(f'Mean Squared Error (Polynomial Regression): {mse_poly:.2f}')
```
